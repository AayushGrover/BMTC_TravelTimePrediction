```python
import pandas as pd
import glob
import ntpath
```

```python
%store -r PATH_TO_DATA_FOLDER

ind_files = PATH_TO_DATA_FOLDER+"/data"
```

```python
for filepath in glob.iglob(ind_files+"/*.csv"):
    data=pd.read_csv(filepath)
#     print(ntpath.basename(filepath))
    data = data.sort_values(by="time")
    data.to_csv(filepath,index=None)
    print(data.head())
```
