```python
import pandas as pd
import glob
import os
import datetime
import pickle
import ntpath
```

```python
%store -r PATH_TO_DATA_FOLDER

ind_files = PATH_TO_DATA_FOLDER+"/data"

with open("dictionary.pickle","rb") as input_file:
    dict_buses = pickle.load(input_file,encoding="latin1")

count_t = 0    
count_f = 0
count_e = 0

print(dict_buses[150218000])

for i in dict_buses:
    if dict_buses[i] == True:
        count_t += 1
    elif dict_buses[i] == False:
        count_f += 1
    else:
        count_e += 1
        
print(count_t)
print(count_f)
print(count_e)
    
combined_data = pd.DataFrame(columns=["lat1","long1","lat2","long2","time_taken","date","hours","minutes","seconds"])
# combined_data = pd.DataFrame(columns=["lat1","long1","lat2","long2","time_taken",,"hours","minutes","seconds"])

combined_data
```

```python
#iterrows to get each row of the dataframe
class CombinedRow:
    def __init__(self,row1,row2):
        self.lat1 = row1["lat"]
        self.long1 = row1["long"]
        self.lat2 = row2["lat"]
        self.long2 = row2["long"]
        
        self.time_list_1 = row1["time"].split()[1].split(":")
        self.time_list_2 = row2["time"].split()[1].split(":")
        self.timestamp1 = datetime.datetime(year=2016,month=7,day=2,hour=int(self.time_list_1[0]),minute=int(self.time_list_1[1]),second=int(self.time_list_1[2]))
        self.timestamp2 = datetime.datetime(year=2016,month=7,day=2,hour=int(self.time_list_2[0]),minute=int(self.time_list_2[1]),second=int(self.time_list_2[2]))
    
        self.time_taken = (self.timestamp2 - self.timestamp1).total_seconds()
        #gives the time difference in seconds
        
        self.date = row1["time"].split()[0]
        self.hours = self.timestamp1.hour
        self.minutes = self.timestamp1.minute
        self.seconds = self.timestamp1.second
        
    def getDictToAdd(self):
#         return [[self.lat1,self.long1,self.lat2,self.long2,self.time_taken,self.date,self.hours,self.minutes,self.seconds]]
        return [{"lat1":self.lat1,"long1":self.long1,"lat2":self.lat2,"long2":self.long2,"time_taken":self.time_taken,"date":self.date,"hours":self.hours,"minutes":self.minutes,"seconds":self.seconds}]
```

```python
i = 0
for filepath in glob.iglob(ind_files+"/*.csv"):
    data=pd.read_csv(filepath)
    bus_no = int(ntpath.basename(filepath).split(".")[0])
    
    #since there are more false buses, thats the training set
    try:
        if(dict_buses[bus_no] == True):
            continue
    except:
        print("bus number "+ str(bus_no) + " not found")
        
        #these buses are also included in the dataset
    row1 = data.iloc[0]
    for index,row in data.iterrows():
        if index == 0:
            continue
        
        new_row = CombinedRow(row1,row).getDictToAdd()
        combined_data = combined_data.append(new_row)
        row1 = row

    if i == 0:
        combined_data.to_csv(PATH_TO_DATA_FOLDER + "/combined_data.csv",index=None)
    else:
        combined_data.to_csv(PATH_TO_DATA_FOLDER + "/combined_data.csv",index=None,mode = 'a',header= None)
    
    combined_data = pd.DataFrame(columns=["lat1","long1","lat2","long2","time_taken","date","hours","minutes","seconds"])
        
    i+=1
print(i)
```
